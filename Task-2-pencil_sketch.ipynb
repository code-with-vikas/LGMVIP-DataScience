```python
import cv2
import matplotlib.pyplot as plt
```

```python
image =cv2.imread("dog.jpg")
```

# convert Bgr ro rgb 

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Dog',size=25)
plt.axis(False)
plt.show()
```

# Image Processing

Step:1--RGB to Grey image

```python
grey_img=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
plt.title('gray',size=25)
plt.axis(False)
plt.show()
```

Step-2-Inverting_image

```python
invert_img=cv2.bitwise_not(grey_img)
plt.imshow(invert_img)
plt.title('inverted grey',size=25)
plt.axis(False)
plt.show()
```

step-3-blurring image

```python

blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(blur_img)
plt.title('burred',size=25)
plt.axis(False)
plt.show()
```

STep-4-invetring blurred image

```python
invblur_img=cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
plt.title('inverted_blurred',size=25)
plt.axis(False)
plt.show()
```

Step-5-Guassianblurring

```python
blur = cv2.GaussianBlur(invblur_img, (21, 21), 0)
plt.imshow(blur)
plt.title('gaussian_blurred',size=25)
plt.axis(False)
plt.show()
```

```python
invertedblur = cv2.bitwise_not(blur)
plt.imshow(invertedblur)
plt.title('inverted_gaussian_blurred',size=25)
plt.axis(False)
plt.show()
```

```python
sketch_img=cv2.divide(grey_img,invertedblur, scale=256.0)
```

```python
rgb_sketch=cv2.cvtColor(sketch_img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.title('sketch_image',size=25)
plt.axis(False)
plt.show()
```
